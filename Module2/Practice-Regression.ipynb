```python
def clean_auto_data(data):
    
    '''
    Function to clean data
    
    Arguments
    ---------
    
    data -> pd.dataframe: dataframe to be cleaned
    
    Returns
    -------
    
    cleaned dataframe object
    '''
    
    # replace '-' with '_'
    new_column_names = []
    for column in data.columns:
        column = column.replace('-', '_')
        new_column_names.append(column)
        
    data.columns = new_column_names # replacing column names
    
    # replacing '?' with numpy "Not a Value"
    for column in data.columns:
        data[column].loc[data[column] == '?'] = np.nan
        
    columns = ['price', 'bore', 'stroke', 'horsepower', 'peak_rpm'] # columns to change data type to numeric

    for column in columns:
        data[column] = pd.to_numeric(data[column])
        
    return data
```

```python
data = pd.read_csv('Automobile price data _Raw_.csv')  # loading data with pandas

data = clean_auto_data(data)
```

```python
def check_unique_cat_counts(data):
    
    for column in data.columns:
        if str(data[column].dtype) == 'object':
            print(f'For column {column} category counts are:')
            print(data[column].value_counts())
            print(f'--------------------')
```

```python
def plot_bars(data, cols):
    fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(12,30))
    
    for i in range(len(cols)):
        
        ax[i].grid()   # to remove grid lines
        
        counts = dict(data[cols[i]].value_counts())
        labels = counts.keys()
        values = counts.values()
        ax[i].bar(labels, values)
        ax[i].set_title(f'Plot of {cols[i].capitalize()} against Frequency')
        ax[i].set_ylabel('Frequency')
        ax[i].set_xlabel(f'{cols[i]}')

plot_cols = ['make', 'body_style', 'num_of_cylinders']
plot_bars(data, plot_cols)    
```

```python
def plot_hist(data, cols):
    fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(18,18))
    
    for i in range(len(cols)):
        
        counts = dict(data[cols[i]].value_counts())
        labels = counts.keys()
        values = counts.values()
        #ax[i].hist(values, bins=10)
        plt.hist(values, bins=10)
        plt.show()

num_cols = ['curb_weight', 'engine_size', 'city_mpg', 'price']    
plot_hist(data, num_cols)
```

```python
def plot_histogram(auto_prices, cols, bins = 10):
    for col in cols:
        fig = plt.figure(figsize=(6,6)) # define plot area
        ax = fig.gca() # define axis    
        auto_prices[col].plot.hist(ax = ax, bins = bins) # Use the plot.hist method on subset of the data frame
        ax.set_title('Histogram of ' + col) # Give the plot a main title
        ax.set_xlabel(col) # Set text for the x axis
        ax.set_ylabel('Number of autos')# Set text for y axis
        plt.show()
        
num_cols = ['curb_weight', 'engine_size', 'city_mpg', 'price']    
plot_histogram(data, num_cols)
```

```python
def show_density_hist(data, columns, bins, hist=True):
    
    '''
    function to show density histogram
    
    Argumnts
    ---------
    data: dataframe object
    columns: numeric columns 
    bins
    hist: specify True to display histogram plot, False for otherwise
    '''
    
    sns.set_style("whitegrid")
    for column in columns:
        
        fig = plt.figure(figsize=(10,6))
        sns.distplot(data[column], bins=bins, rug=True, hist=hist)
        plt.title('Histogram of ' + column) # Give the plot a main title
        plt.xlabel(column) # Set text for the x axis
        plt.ylabel('Number of autos')# Set text for y axis
        plt.show()    
```

```python
show_density_hist(data, num_cols, 10, hist=False)
```

```python
show_density_hist(data, num_cols, 10, hist=True)
```

## 2D Plots

```python
def scatter_plots(data, cols, label):
    fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(8,24))
    
    for i in range(len(cols) - 1):
        
        ax[i].grid()
        counts = dict(data[cols[i]].value_counts())
        labels = counts.keys()
        values = counts.values()
        ax[i].scatter(data[cols[i]], data[label])
        ax[i].set_title(f'Plot of {cols[i].capitalize()} against Price')
        ax[i].set_ylabel('Price')
        ax[i].set_xlabel(f'{cols[i]}')

plot_cols = ['curb_weight', 'engine_size', 'horsepower', 'city_mpg']
scatter_plots(data, plot_cols, 'price')    
```

### Overplotting

```python
def plot_2d_density(auto_prices, columns, label='price', kind ='kde'):
    for column in columns:
        sns.set_style("whitegrid")
        sns.jointplot(column, label, data=auto_prices, kind=kind)
        plt.xlabel(column) # Set text for the x axis
        plt.ylabel(label)
        plt.show()

plot_2d_density(data, num_cols)      
```

### Conmparing relationships between categorical and numerical variables

#### Box Plots

```python
def plot_box(data, cols, label='price'):
    for col in cols:
        #sns.set_style("whitegrid")
        fig = plt.figure(figsize=(12, 8))
        sns.boxplot(col, label, data=data)
        plt.xlabel(col) 
        plt.ylabel(label)
        plt.show()
        
cat_cols = ['fuel_type', 'aspiration', 'num_of_doors', 'body_style', 
            'drive_wheels', 'engine_location', 'engine_type', 'num_of_cylinders']
```

```python
plot_box(data, cat_cols)    
```

```python
def plot_violins(data, cols, label='price'):
    for col in cols:
        #sns.set_style("whitegrid")
        fig = plt.figure(figsize=(12, 8))
        sns.violinplot(col, label, data=data)
        plt.xlabel(col) 
        plt.ylabel(label)
        plt.show()
```

```python
plot_violins(data, cat_cols)
```
